```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
plt.style.use('ggplot')
```

```python
data = pd.read_csv('dataset/nmesdata.txt')
data = data[["LASTAGE", "TOTALEXP", "marital", "AGESMOKE", "educate", "beltuse", "MALE", "RACE3"]].dropna()
data["LASTAGE"] = (data["LASTAGE"]-data["LASTAGE"].mean())/data["LASTAGE"].std()
data["TOTALEXP"] = (data["TOTALEXP"]-data["TOTALEXP"].mean())/data["TOTALEXP"].std()
data["marital"] = (data["marital"] == 1).astype(float)
data["marital"] = (data["marital"]-data["marital"].mean())/data["marital"].std()
```

```python
sns.jointplot(data["LASTAGE"], data["TOTALEXP"])
```

```python
sns.jointplot(data["marital"], data["TOTALEXP"])
```

```python
np.cov(data['marital'], data['TOTALEXP'])
```
