```python
import numpy as np
import numpy.random as npr
import seaborn as sns
import matplotlib.pyplot as plt
plt.style.use('ggplot')
```

```python
X = np.arange(-5, 5, 0.01)
noise = npr.normal(0,1, len(X))*(X**2-3*X)
base = 5*X**2 -X
Y = base + noise
Y_ = base
noise = noise
```

```python
fig, ax = plt.subplots(1, 2, figsize=(12,4))
sns.lineplot(X, Y, alpha=0.8, ax=ax[0])
sns.lineplot(X, Y_, ax=ax[0])
ax[0].set_title('data')
sns.lineplot(X, noise, ax=ax[1], alpha=0.8)
ax[1].set_title('noise')
```

```python
X = np.arange(-5, 5, 0.01)
noise = npr.normal(0,1, len(X))*1/(np.maximum(np.abs(X), 0.3))
base = 1/2*X**2 + 3*np.sin(2*X)-X
Y = base + noise
Y_ = base
noise = noise
```

```python
fig, ax = plt.subplots(1, 2, figsize=(12,4))
sns.lineplot(X, Y, alpha=0.8, ax=ax[0])
sns.lineplot(X, Y_, ax=ax[0])
ax[0].set_title('data')
sns.lineplot(X, noise, ax=ax[1], alpha=0.8)
ax[1].set_title('noise')
```
